# Data Balancing usng SMOTE, Data Augmentation

```python
# Separate the Feature Variables (X) and Target/ Class Variables (y)
X = df.drop(columns=['Attack Type'])  # Features
y = df['Attack Type']  # Target

# Print the dimension of the dataset
print("Dimension of the Dataset:", df.shape)

# Print the dimensions of Feature Variables (X) and Target Variable (y)
print("Dimension of Feature Variables (X):", X.shape)
print("Dimension of Target Variable (y):", y.shape)
```

```python
import numpy as np
import pandas as pd
from imblearn.over_sampling import SMOTE

# Set the target size for data balancing
target_size = 200000

# Count the number of samples needed to balance the classes
class_counts = y.value_counts().to_dict()
total_samples_needed = target_size - len(y)

# Calculate how many samples should be added for each class
samples_per_class = {cls: count + total_samples_needed // len(class_counts) for cls, count in class_counts.items()}

# Apply SMOTE for data balancing
smote = SMOTE(sampling_strategy=samples_per_class, random_state=42)
X_res, y_res = smote.fit_resample(X, y)

# Function to add noise to the data for augmentation
def add_noise(dataframe, noise_level=0.01):
    noise = np.random.randn(*dataframe.shape) * noise_level
    dataframe_noisy = dataframe + noise
    return dataframe_noisy

# Select numerical columns for noise addition
numerical_columns = X_res.select_dtypes(include=[np.number]).columns

# Apply noise to numerical columns
X_res_noisy = add_noise(X_res[numerical_columns])

# Augment the dataset with noisy samples
X_augmented = pd.concat([X_res, X_res_noisy], ignore_index=True)
y_augmented = pd.concat([y_res, y_res], ignore_index=True)

# Combine the augmented features and target variable into one DataFrame
df_resampled = X_augmented.copy()
df_resampled['Attack Type'] = y_augmented

# Display the info of the resampled dataset
df_resampled.info()

# Print the number of unique attack classes and their counts
print("Number of Attack Classes:", df_resampled['Attack Type'].unique())
print("\nNumber of Samples for each Attack Class:", df_resampled['Attack Type'].value_counts())
```

```python
import matplotlib.pyplot as plt

# Plot the distribution of attacks after data balancing using SMOTE
plt.figure(figsize=(10,5))
df_resampled['Attack Type'].value_counts().plot(kind='bar')

# Setting custom x-ticks to represent the attack types
plt.xticks(ticks=[0, 1, 2], labels=['DDoS', 'Intrusion', 'Malware'], rotation=0)

# Label the y-axis and set the title
plt.ylabel('No. of Attacks')
plt.title("Distribution of Attacks (After Data Balancing)")

# Show the plot
plt.show()
```

# Data Standardization or Normalization

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.utils import to_categorical
import numpy as np

# Initial split: Train and Test
X_train, X_test, y_train, y_test = train_test_split(
    df_resampled.drop(columns=['Attack Type']), 
    df_resampled['Attack Type'], 
    test_size=0.2,  # 20% for testing
    random_state=42
)

# Further split the training set into Training and Validation sets
X_train, X_val, y_train, y_val = train_test_split(
    X_train, y_train, 
    test_size=0.2,  # 20% of the training set for validation
    random_state=42
)

# Feature scaling using StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_val = scaler.transform(X_val)
X_test = scaler.transform(X_test)

# Convert labels to categorical format (One-hot encoding)
num_classes = len(np.unique(y_train))
y_train = to_categorical(y_train, num_classes)
y_val = to_categorical(y_val, num_classes)
y_test = to_categorical(y_test, num_classes)

# Print the dimensions of the datasets
print("After Feature Scaling and Splitting into Train, Validation, and Test sets:")
print("Dimensions of TRAINING SET:", X_train.shape, y_train.shape)
print("Dimensions of VALIDATION SET:", X_val.shape, y_val.shape)
print("Dimensions of TESTING SET:", X_test.shape, y_test.shape)

# Display the first 2 rows of each set
print("\nFirst 2 rows of X_train:")
print(X_train[:2])

print("\nFirst 2 rows of y_train:")
print(y_train[:2])

print("\nFirst 2 rows of X_val:")
print(X_val[:2])

print("\nFirst 2 rows of y_val:")
print(y_val[:2])

print("\nFirst 2 rows of X_test:")
print(X_test[:2])

print("\nFirst 2 rows of y_test:")
print(y_test[:2])
```
